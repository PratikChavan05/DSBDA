```python
import pandas as pd

nba = pd.read_csv("nba.csv")
nba.head()
```

```python
nba.info()
```

```python
nba.isnull().sum()
```

```python
nba["Salary"] = nba["Salary"].fillna(nba["Salary"].mean())
nba["Age"] = nba["Age"].fillna(nba["Age"].mean())
```

```python
grouped = nba.groupby("Team")["Salary"]
# This creates a list of all salaries belonging to each team
salary_lists = grouped.apply(list)
print(salary_lists)
```

```python
summary = grouped.agg(['mean', 'median', 'min', 'max', 'std'])
summary
```

```python
salary_list = nba.groupby("Team")["Salary"].apply(list)
salary_list
```
