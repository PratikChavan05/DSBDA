```python
with open("txt1.txt") as file:
    doc1=file.read()

with open("txt2.txt") as file:
    doc2=file.read()

    
```

```python
from nltk.tokenize import word_tokenize , sent_tokenize

# Sentence Level Tokenization
senttok = sent_tokenize(doc1)


#Word level tokenization
wordtok = word_tokenize(doc1)
```

```python

from nltk.corpus import stopwords

##################################################################################

# 🔍 What is stopwords?

# A predefined list of common useless words

# Examples:
# the, is, in, and, are, of, to, a

##################################################################################


stop_words = set(stopwords.words('english'))
list(stop_words)[:10]
```

```python
wordtok = [token for token in wordtok if token not in stop_words]
wordtok[:5]
```

```python
# 🔷 2. POS TAGGING
# 👉 Definition:

# Assign grammatical role to each word.
# nouns ('NN')
# verb ('VBZ')
# adjectives ('JJ')
# determiners ('DT')

pos_tags = nltk.pos_tag(wordtok)
pos_tags[:10]
```

```python
# 🔷 4. STEMMING
# 👉 Definition:

# Cut words to root form (may not be correct English)

# Example:
# playing → play
# studies → studi
# running → run

# 🔍 Types of Stemmers:
# Porter (most common)
# Lancaster (aggressive)
# Snowball


from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmedwords=[]
for word in wordtok:
    stemmedwords.append(stemmer.stem(word))

dict(zip(wordtok[:10],stemmedwords[:10]))
```

```python
# 🔷 5. LEMMATIZATION
# 👉 Definition:

# Convert word → dictionary base form

# Example:
# running → run
# better → good

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemwords=[]
for word in wordtok:
    #lemwords.append(lemmatizer.lemmatize(word))

    # ⚠ Important:
    # Better results if POS given:
     lemwords.append(lemmatizer.lemmatize(word, pos='v'))

dict(zip(wordtok[:50],lemwords[:50]))
```

```python
# 🔷 6. TERM FREQUENCY (TF)
# 👉 Definition:

# How often word appears in document

# Formula:
# TF = count(word) / total words

#----------------------------------------------------------------------------------------------

##### USING LIBRARY ----- Calculating frequency 

from nltk.probability import FreqDist
fdist = FreqDist(wordtok)
fdist

# total number of words
total_words = len(wordtok)

# calculate TF
tf = {}

for word, count in fdist.items():
    tf[word] = count / total_words
# example
print(tf)

#----------------------------------------------------------------------------------------------

##### USING Counter  ----- Calculating frequency 

from collections import Counter
fdist1 = Counter(wordtok)

total_words = len(wordtok)
tf = {}

for word, count in fdist1.items():
    tf[word] = count / total_words

print(tf)

#----------------------------------------------------------------------------------------------



```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Step 2: Create vectorizer
vectorizer = TfidfVectorizer()

# Step 3: Fit and transform
tfidf_matrix = vectorizer.fit_transform([doc1,doc2])

# Step 4: Get words (features)
feature_names = vectorizer.get_feature_names_out()

# Step 5: Convert to array (for display)
print("Words:", feature_names)
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

# TF
cv = CountVectorizer()
tf = cv.fit_transform([doc1,doc2])

print("Term Frequency:\n", tf.toarray())

# IDF
tfidf_transformer = TfidfTransformer()
tfidf_transformer.fit(tf)

print("IDF:\n", tfidf_transformer.idf_)
```
