```python
import nltk
import math
import string
```

```python
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize  import word_tokenize ,  sent_tokenize
from nltk.stem import PorterStemmer , WordNetLemmatizer 
from sklearn.feature_extraction.text  import TfidfVectorizer
```

```python
nltk.download("punkt")
nltk.download("wordnet")
nltk.download("stopwords")
nltk.download("averaged_perceptron_tagger")
```

```python
with open("txt1.txt") as f:
    doc1 = f.read()

with open("txt2.txt") as f:
    doc2 = f.read()
```

```python
def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('','', string.punctuation))
    words = word_tokenize(text)

    s_w = set(stopwords.words('english'))
    words = [w for w in words if w not in s_w]
    return words

token1 = preprocess(doc1)
token2 = preprocess(doc2)
```

```python
sent1  = sent_tokenize(doc1)
print("sent1:")
for s  in sent1[:3]:
    print(s)
    print("-------------")

sent2  = sent_tokenize(doc1)
print("sent2:")
for s  in sent2[:3]:
    print(s)
    print("-------------")
```

```python
p_t = nltk.pos_tag(token1)
print(p_t[:10])
```

```python
stemmer  = PorterStemmer()
stemmed  = [stemmer.stem(w) for w in token1]
print(dict(zip(token1[:10],stemmed[:10])))
```

```python
lemma  = WordNetLemmatizer()
lemmed = [lemma.lemmatize(w , pos ='v') for w in token1]
print(dict(zip(token1[:10], lemmed[:10])))
```

```python
def comp_tf(token):
    tf ={}
    total = len(token)
    freq = Counter(token)
    for w , c  in freq.items():
        tf[w] = c/total

    return tf

tf1 = comp_tf(token1)
tf2 = comp_tf(token2)
print(list(tf1.items())[:10])
print("-"*100)
print("tf2:", list(tf2.items())[:10])
```

```python
a_w  = set(token1) | set(token2)
N=2

idf = {}

for w in a_w:
    df = (1 if w in token1 else 0)+(1 if w in token2 else 0)
    idf[w] = math.log(N/df)

print(list(idf.items())[:10])
```

```python
vec =  TfidfVectorizer(lowercase = True,  stop_words='english')
tfidf1 = vec.fit_transform([doc1,doc2])
f_n = vec.get_feature_names_out()
print(f_n[:10])
print(tfidf1.toarray())
```

```python
tfidf1 = {}
tfidf2 = {}

for word in a_w:
    tfidf1[word] = tf1.get(word, 0) * idf[word]
    tfidf2[word] = tf2.get(word, 0) * idf[word]

print("\nTF-IDF (Doc1 Sample):", list(tfidf1.items()))
```
