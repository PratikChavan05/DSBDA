```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df.drop(columns=["Cabin","SibSp"])
```

```python
pratik = df.drop_duplicates()
df["Age"].nunique()
```

```python
plt.figure()
sns.boxplot(x="Fare" ,data=df , hue = "Sex",  showfliers = False)
plt.title("Fare X Survived")
plt.show()
```

```python
plt.figure()
sns.countplot(x="Fare",data=df , hue = "Sex")
plt.title("Fare X Survived")
plt.show()
```

```python
plt.figure()
sns.histplot(x=df['Age'], hue =  df['Sex'] , kde= True , bins= 30)
plt.title("Fare X Survived")
plt.show()
```
