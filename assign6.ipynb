```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = pd.read_csv("Iris.csv")   # or sns.load_dataset("iris")

iris.head()
```

```python
iris.info()
```

```python
# ==============================
# Data Preprocessing (Iris)
# ==============================

# 1. Check missing values
print("Missing Values:\n", iris.isnull().sum())

# 2. Handle missing values (if any)
iris = iris.dropna()   # Iris usually has no missing values

# 3. Check duplicates
print("Duplicate Rows:", iris.duplicated().sum())

# Remove duplicates (if any)
iris = iris.drop_duplicates()

# 4. Check data types
print("\nData Types:\n", iris.dtypes)

# 5. Drop unnecessary column (if present)
iris.drop(columns=["Id"], inplace=True)

# 6. Check outliers using IQR (optional but good for marks)
for col in iris.select_dtypes(include=['float64', 'int64']).columns:
    Q1 = iris[col].quantile(0.25)
    Q3 = iris[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Capping outliers
    iris[col] = iris[col].clip(lower, upper)

print("\nPreprocessing Completed")
```

```python
for col in iris.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure()
    sns.boxplot(y=iris[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
for col in iris.select_dtypes(include=["int64", "float64"]).columns:
    plt.figure()
    sns.histplot(data=iris , x = col , kde = True , bins = 30)
    plt.title(f"Histplot of {col}")
    plt.show()
```
