```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag
from nltk.stem  import PorterStemmer , WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
            
```

```python
with open("txt1.txt") as f:
    doc1 = f.read()

with open("txt2.txt") as f:
    doc2 = f.read()

doc = doc1+ " "+doc2
```

```python
tokens = word_tokenize(doc)
print(tokens)
```

```python
pt = pos_tag(tokens)
print(pt)
```

```python
sw = set(stopwords.words('english'))
fd = [w for w in tokens if w not in sw and w.isalpha()]
print(fd)
```

```python
stemmer  = PorterStemmer()
stemmed  = [stemmer.stem(w) for w in  fd]
print(stemmed)
```

```python
lem  = WordNetLemmatizer()
lemed  = [lem.lemmatize(w) for w in  fd]
print(lemed)
```

```python
doc3  = [doc1, doc2]
```

```python
vec = TfidfVectorizer()
tm = vec.fit_transform(doc3)
print(vec.get_feature_names_out())
print(tm.toarray())
```
